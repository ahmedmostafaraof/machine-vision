```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
img1 = cv2.imread('median.png')
img_rgb = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img_blurred = cv2.medianBlur(img_rgb, (5))
titles = ['Original Image', 'median Blurred Image']
images = [img_rgb, img_blurred]
plt.figure(figsize=(5, 3))
for i in range(2):
    plt.subplot(1, 2, i + 1)
    plt.imshow(images[i])
    plt.title(titles[i], fontsize=10)
    plt.xticks([]), plt.yticks([])
plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
img1 = cv2.imread('wood.png')
img_rgb = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img_blurred = cv2.bilateralFilter(img_rgb, d=15, sigmaColor=75, sigmaSpace=75)
titles = ['Original Image', 'Bilateral Blurred Image']
images = [img_rgb, img_blurred]
plt.figure(figsize=(5, 3))
for i in range(2):
    plt.subplot(1, 2, i + 1)
    plt.imshow(images[i])
    plt.title(titles[i], fontsize=10)
    plt.xticks([]), plt.yticks([])
plt.tight_layout()
plt.show()
```
