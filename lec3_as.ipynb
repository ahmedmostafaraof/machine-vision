```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


img = np.zeros((500, 500, 3))


cv2.rectangle(img, (100, 100), (150, 200), (255, 0, 0), 2)  
cv2.rectangle(img, (200, 200), (300, 300), (0, 255, 0), 2)  
cv2.rectangle(img, (250, 320), (400, 400), (0, 0, 255), 2)  


cv2.imshow('Rectangles', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np


img = cv2.imread(r'my_id.jpg')
cv2.putText(img, "ahmed", (500, 500), cv2.FONT_HERSHEY_COMPLEX, 3, (200, 100, 255), 5)
((text_width,text_height),baseline)=cv2.getTextSize("ahmed", cv2.FONT_HERSHEY_COMPLEX,3, 5)
cv2.line(img, (500,510), (500+text_width, 510),(0,255,0), 2)
height, width = img.shape[:2]


corners = [
    (0, 0),                
    (width-1 , 0),        
    (0, height-1 ),       
    (width-1 , height-1 ) 
]

for point in corners:
    cv2.circle(img, point, 100, (255, 0, 0), 3)
cv2.imshow('my_id', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
just_id=cv2.imread(r'my_id.jpg')
img=cv2.cvtColor(just_id, cv2.COLOR_BGR2RGB)
alpha_values = [1.0, 0.75, 0.5, 0.25]

plt.figure(figsize=(12, 6))
for i, alpha in enumerate(alpha_values):
    plt.subplot(1, 4, i+1)  
    plt.imshow(img, alpha=alpha)
    plt.title(f'Alpha = {alpha}')
    plt.axis('off')
plt.show()
```
